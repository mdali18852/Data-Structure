```python
################################## DATA STRUCTURE SHEET-1 #######################################

# Q.1 Write a code to reverse a string
'''
st="Physics Wallah"
reversed_st = st[::-1]
print("This is original string: ",st)
print("Reverse os string is: ",reversed_st)
print("-"*50,"\n")
                                          #Using function
def reverse_st():
    name="Alimullah"
    print("Without reverse name: ",name)
    reverse=name[::-1]
    return reverse
call=reverse_st()
print("Revese of name: ",call)
print("-"*50,"\n")
'''

# Q.2 Write a code to count the number of vowels in a string
'''
def count_def():
    name="Muzaffarpur, Darbhanga, Patna"
    count=0
    for i in name:
        if i in 'aeiouAEIOU':
            count+=1
    return count

print("Number of vowels in name: ",count_def())

s="Alimullah"
vowels="aeiouAEIOU"
count=0
for char in s:
    if char in vowels:
        count+=1
print("Number of vowels in name: ",vowels)
'''

# Q.3 Write a code to check if a given string is a palindrome or not
'''
st=input("Enter the String name: ")
if st==st[::-1]:
    print("This is the polindrome string")
else:
    print("This is not polindrome string: ")
'''
#Q.4 Write a code to check if two given strings are anagrams of each other

#Both strings have the same characters: l, i, s, t, e, n (frequency of each character is the same).
#Rearranging the characters of "listen" gives "silent", so they are anagrams.
'''
str1=input("Enter the first String: ")
str2=input("Enter the 2nd String: ")
if sorted(str1)==sorted(str2):
 print("This is anagrams string: ")
else:
   print("This is not anagrams string: ")
'''

#Q.5 Write a code to find all occurrences of a given substring within another string
'''
def find_substring_occurrences(main_string, substring):
    positions = []
    start = 0

    while start < len(main_string):
        start = main_string.find(substring, start)

        if start == -1:
            break

        positions.append(start)
        start += 1  # move past the last match to continue searching

    return positions

# Example usage
main_string = "This is a test string, and this is a test."
substring = "test"
occurrences = find_substring_occurrences(main_string, substring)
print(f"Occurrences of '{substring}' are at positions: {occurrences}")
'''

#Q.6 Write a code to perform basic string compression using the counts of repeated characters
'''
def compress_string(s):
    compressed = []
    count = 1

    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1

    compressed.append(s[-1] + str(count))  # Add the last character and its count
    compressed_str = ''.join(compressed)

    return compressed_str if len(compressed_str) < len(s) else s  # Return original if no compression

# Example usage
text = input("Enter a string: ")
print("Compressed string:", compress_string(text))

#Q.7 Write a code to determine if a string has all unique characters
def is_unique(s):
    return len(set(s)) == len(s)

text = input("Enter a string: ")
print("All characters are unique:", is_unique(text))
'''

#Q.8 Write a code to convert a given string to uppercase or lowercase
'''
text = input("Enter a string: ")
choice = input("Convert to Uppercase or Lowercase? ").strip().lower()

if choice == 'u':
    print("Uppercase:", text.upper())
elif choice == 'l':
    print("Lowercase:", text.lower())
else:
    print("Invalid choice!")
'''
#Q.9 Write a code to count the number of words in a string
'''
text = input("Enter a string: ")
words = text.split()  # Splitting the string by spaces
count = len(words)  # Counting the number of words
print("Number of words:", count)
'''

#Q.10 Write a code to concatenate two strings without using the + operator

'''
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

concatenated = "{}{}".format(str1, str2)  # Using format() method
print("Concatenated string:", concatenated)
'''

#Q.11 Write a code to remove all occurrences of a specific element from a list
'''
lst = [int(x) for x in input("Enter list elements separated by space: ").split()]
element = int(input("Enter the element to remove: "))

filtered_list = [x for x in lst if x != element]  # List comprehension to remove element

print("Updated list:", filtered_list)
'''
#Q.12 Implement a code to find the second largest number in a given list of integers
'''
lst = list(map(int, input("Enter list elements separated by space: ").split()))

unique_lst = list(set(lst))  # Remove duplicates
unique_lst.sort(reverse=True)  # Sort in descending order

if len(unique_lst) < 2:
    print("No second largest number exists.")
else:
    print("Second largest number:", unique_lst[1])
'''
#Q. 13 Create a code to count the occurrences of each element in a list and return a dictionary with elements as
#keys and their counts as values

'''
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
count_dict = {}  # Empty dictionary to store counts

for num in numbers:
    if num in count_dict:
        count_dict[num] += 1  # Increment count if number already exists
    else:
        count_dict[num] = 1  # Initialize count if number is new

print("Occurrences:", count_dict)
'''
#Q.14 Write a code to reverse a list in-place without using any built-in reverse functions
'''
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

# Using two-pointer approach
left, right = 0, len(numbers) - 1

while left < right:
    numbers[left], numbers[right] = numbers[right], numbers[left]  # Swap elements
    left += 1
    right -= 1

print("Reversed list:", numbers)
'''

#Q. 15 Implement a code to find and remove duplicates from a list while preserving the original order of
#elements
'''
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
unique_list = []  # List to store unique elements

for num in numbers:
    if num not in unique_list:
        unique_list.append(num)  # Add only if not already present

print("List after removing duplicates:", unique_list)
'''

#Q.16 Create a code to check if a given list is sorted (either in ascending or descending order) or not
'''
lst=[2,4,5,7,8,9,10]
if lst == sorted(lst):
    print("List is sorted in ascending order: ",lst)
elif lst == sorted(lst,reverse=True):
    print("List is sorted in decending order: ",lst)
else:
    print("List is not sorted: ",lst)
'''

#Q.17 Write a code to merge two sorted lists into a single sorted list
'''
lst1=[1,6,9,4,5]
lst2=[6,7,8,9,10]
# Using list comprehension and sorted function
merged_list = sorted(lst1+lst2)
print("Merged list: ",merged_list)
'''
#Q. 18 Implement a code to find the intersection of two given lists
'''
lst1=[2,4,7,9,3]
lst2=[1,0,3,4,6]
# Using set intersection operation
intersection_set = set(lst1) or set(lst2)
print("Intersection of two lists: ",set(intersection_set)) # 3,4
'''

#Q.19 Create a code to find the union of two lists without duplicates
'''
lst1=[2,3,5,6,7]
lst2=[6,4,7,8,9]

# Using set union operation
union_set = set(lst1) | set(lst2)
print("Union of two lists: ",set(union_set))
'''

#Q. 20 Write a code to shuffle a given list randomly without using any built-in shuffle functions

#Q.21 Write a code that takes two tuples as input and returns a new tuple containing elements that are
#common to both input tuples

'''
def common_elements(tuple1, tuple2):
    # Convert the tuples to sets and find the intersection
    common = set(tuple1).intersection(set(tuple2))
    # Convert the set back to a tuple and return it
    return tuple(common)

# Example usage:
tuple1 = (1, 2, 3, 4, 5)
tuple2 = (4, 5, 6, 7, 8)

result = common_elements(tuple1, tuple2)
print(result)  # Output: (4, 5)
'''

#Q.22 Create a code that prompts the user to enter two sets of integers separated by commas. Then, print the
#intersection of these two sets

'''
def get_integer_set(prompt):
    # Prompt the user for input and convert it to a set of integers
    user_input = input(prompt)
    # Split the input by commas, strip whitespace, and convert to integers
    integer_set = set(map(int, map(str.strip, user_input.split(','))))
    return integer_set

def main():
    # Prompt the user for the first set of integers
    set1 = get_integer_set("Enter the first set of integers separated by commas: ")
    # Prompt the user for the second set of integers
    set2 = get_integer_set("Enter the second set of integers separated by commas: ")

    # Compute the intersection of the two sets
    intersection = set1.intersection(set2)

    # Print the result
    print("The intersection of the two sets is:", intersection)

# Run the program
if __name__ == "__main__":
    main()
'''
# Q.23 Write a code to concatenate two tuples. The function should take two tuples as input and return a new
#tuple containing elements from both input tuples.
'''
def concatenate_tuples(tuple1, tuple2):
    # Concatenate the two tuples using the + operator
    concatenated_tuple = tuple1 + tuple2
    return concatenated_tuple

# Example usage:
tuple_a = (1, 2, 3)
tuple_b = (4, 5, 6)
result = concatenate_tuples(tuple_a, tuple_b)
print("Concatanate of two tuples: ",result)  # Output: (1, 2, 3, 4, 5, 6)
'''

#Q.24  Develop a code that prompts the user to input two sets of strings. Then, print the elements that are
#present in the first set but not in the second set

'''
# Function to get user input for a set of strings
def input_set(prompt):
    user_input = input(prompt)
    # Split the input string into a list of strings and convert it to a set
    return set(user_input.split())

# Get the first set of strings from the user
set1 = input_set("Enter the first set of strings (separated by spaces): ")

# Get the second set of strings from the user
set2 = input_set("Enter the second set of strings (separated by spaces): ")

# Find elements present in the first set but not in the second set
difference = set1 - set2

# Print the result
print("Elements present in the first set but not in the second set:", difference)
'''
#Q.25 Create a code that takes a tuple and two integers as input. The function should return a new tuple
#containing elements from the original tuple within the specified range of indices
'''
def slice_tuple(original_tuple, start_index, end_index):
    # Use slicing to get the elements within the specified range
    sliced_tuple = original_tuple[start_index:end_index]
    return sliced_tuple

# Example usage:
original_tuple = (10, 20, 30, 40, 50, 60, 70, 80, 90)
start_index = 2
end_index = 6
result = slice_tuple(original_tuple, start_index, end_index)
print(result)  # Output: (30, 40, 50, 60)
'''
#Q. 26 & Write a code that prompts the user to input two sets of characters. Then, print the union of these two sets

'''
# Function to get user input for a set of characters
def input_set(prompt):
    user_input = input(prompt)
    # Convert the input string into a set of characters
    return set(user_input.replace(" ", ""))  # Remove spaces if any

# Get the first set of characters from the user
set1 = input_set("Enter the first set of characters (no spaces): ")

# Get the second set of characters from the user
set2 = input_set("Enter the second set of characters (no spaces): ")

# Compute the union of the two sets
union_set = set1 | set2

# Print the result
print("Union of the two sets:", union_set)
'''
#Q. 27  Develop a code that takes a tuple of integers as input. The function should return the maximum and
#minimum values from the tuple using tuple unpacking
'''
def find_max_min(numbers):
    # Use built-in functions max() and min() to find the maximum and minimum values
    max_value = max(numbers)
    min_value = min(numbers)
    # Return the results as a tuple
    return max_value, min_value

# Example usage:
input_tuple = (34, 12, 89, 5, 67, 23)
max_val, min_val = find_max_min(input_tuple)  # Tuple unpacking
print("Maximum value:", max_val)  # Output: Maximum value: 89
print("Minimum value:", min_val)  # Output: Minimum value: 5
'''

#Q. 28  & Create a code that defines two sets of integers. Then, print the union, intersection, and difference of these
#two sets
'''

set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Compute the union of the two sets
union_set = set1 | set2

# Compute the intersection of the two sets
intersection_set = set1 & set2

# Compute the difference of the two sets (elements in set1 but not in set2)
difference_set = set1 - set2

# Print the results
print("Set 1:", set1)
print("Set 2:", set2)
print("Union of Set 1 and Set 2:", union_set)
print("Intersection of Set 1 and Set 2:", intersection_set)
print("Difference of Set 1 and Set 2 (Set 1 - Set 2):", difference_set)
'''

#Q.29 Create a code that defines two sets of integers. Then, print the union, intersection, and difference of these
#two sets
'''
# Define two sets of integers
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Union of the two sets
union_set = set1 | set2
print("Union:", union_set)

# Intersection of the two sets
intersection_set = set1 & set2
print("Intersection:", intersection_set)

# Difference of the two sets (set1 - set2)
difference_set = set1 - set2
print("Difference (set1 - set2):", difference_set)
'''
#Q. 30 Write a code that takes a tuple and an element as input. The function should return the count of
#occurrences of the given element in the tuple\
'''
def count_occurrences(tup, element):
    # Count the occurrences of the element in the tuple
    return tup.count(element)

my_tuple = (1, 2, 3, 4, 5, 3, 6, 3)
element_to_count = 3

# Call the function and print the result
result = count_occurrences(my_tuple, element_to_count)
print(f"The element {element_to_count} appears {result} times in the tuple.")
'''






```
